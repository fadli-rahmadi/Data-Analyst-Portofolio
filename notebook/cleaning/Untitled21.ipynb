```python
from pathlib import Path
import nbformat as nbf

nb = nbf.v4.new_notebook()

cells = []

cells.append(nbf.v4.new_markdown_cell("""# Data Cleaning - Balaji Fast Food Sales

Notebook ini berisi proses data cleaning untuk dataset penjualan Balaji Fast Food.
Tahapan yang dilakukan:

1. Import library dan load data
2. Memeriksa struktur dataset
3. Mengecek missing values
4. Mengecek data duplikat
5. Menangani missing value pada `transaction_type`
6. Standarisasi format tanggal
7. Optimasi tipe data
8. Validasi hasil cleaning
9. Menyimpan dataset bersih
"""))

cells.append(nbf.v4.new_code_cell("""import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

df = pd.read_csv('Balaji Fast Food Sales (tb & fadli).csv')
df.head()
"""))

cells.append(nbf.v4.new_markdown_cell("## 1. Informasi Dataset"))

cells.append(nbf.v4.new_code_cell("""print('Shape Dataset:', df.shape)
print('\\nInfo Dataset')
df.info()

print('\\nMissing Values')
print(df.isnull().sum())

print('\\nJumlah Duplikasi:', df.duplicated().sum())
"""))

cells.append(nbf.v4.new_markdown_cell("## 2. Menangani Missing Value"))

cells.append(nbf.v4.new_code_cell("""# Melihat distribusi transaction_type sebelum cleaning
df['transaction_type'].value_counts(dropna=False)
"""))

cells.append(nbf.v4.new_code_cell("""# Mengisi missing value berdasarkan proporsi data yang ada
np.random.seed(42)

distribution = df['transaction_type'].dropna().value_counts(normalize=True)

missing_mask = df['transaction_type'].isna()

df.loc[missing_mask, 'transaction_type'] = np.random.choice(
    distribution.index,
    size=missing_mask.sum(),
    p=distribution.values
)

print(df['transaction_type'].isnull().sum())
"""))

cells.append(nbf.v4.new_markdown_cell("## 3. Standarisasi Format Tanggal"))

cells.append(nbf.v4.new_code_cell("""# Mengubah berbagai format tanggal menjadi datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')

print(df['date'].head())
print(df['date'].isnull().sum())
"""))

cells.append(nbf.v4.new_markdown_cell("## 4. Optimasi Tipe Data"))

cells.append(nbf.v4.new_code_cell("""categorical_cols = [
    'item_name',
    'item_type',
    'transaction_type',
    'received_by',
    'time_of_sale'
]

for col in categorical_cols:
    df[col] = df[col].astype('category')

df.info()
"""))

cells.append(nbf.v4.new_markdown_cell("## 5. Validasi Hasil Cleaning"))

cells.append(nbf.v4.new_code_cell("""print('Missing Values Setelah Cleaning')
print(df.isnull().sum())

print('\\nJumlah Duplikasi')
print(df.duplicated().sum())
"""))

cells.append(nbf.v4.new_markdown_cell("## 6. Analisis Outlier"))

cells.append(nbf.v4.new_code_cell("""Q1 = df['transaction_amount'].quantile(0.25)
Q3 = df['transaction_amount'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[
    (df['transaction_amount'] < lower_bound) |
    (df['transaction_amount'] > upper_bound)
]

print('Jumlah Outlier:', len(outliers))
print('Persentase Outlier:', round((len(outliers)/len(df))*100,2), '%')

sns.boxplot(x=df['transaction_amount'])
plt.show()
"""))

cells.append(nbf.v4.new_markdown_cell("## 7. Menyimpan Dataset Bersih"))

cells.append(nbf.v4.new_code_cell("""df.to_csv('Balaji_FastFood_Cleaned.csv', index=False)

print('Dataset berhasil disimpan sebagai Balaji_FastFood_Cleaned.csv')
"""))

nb.cells = cells

output = "/mnt/data/Data_Cleaning_Balaji_FastFood.ipynb"
with open(output, "w", encoding="utf-8") as f:
    nbf.write(nb, f)

print(output)
```
